```python
#imports liberaries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#reads the dataset
df=pd.read_csv("/kaggle/input/salary-dataset-simple-linear-regression/Salary_dataset.csv")
```

```python
#to plot scatter graph of the dataset
plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter plot")
plt.show()
```

```python
#dependent and in dependent variables
X=df[['YearsExperience']]
Y=df['Salary']
```

```python
#for forming regression line
model = LinearRegression()
model.fit(X,Y)#calculates the best fitting relation between x and y
slope=model.coef_[0]#represents coefficient of x in the equation of line
intercept=model.intercept_ 
```

```python
#function format
def linear_regression(X):
    return slope*X + intercept
```

```python
#to plot the most fitted regression line
plt.scatter(df['YearsExperience'],df['Salary'], label='data points')
plt.plot(df['YearsExperience'],[linear_regression(xi) for xi in df['YearsExperience']], color= 'red', label='Regression Line')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Linear Regression")
plt.show()
```

```python
#prediction for new values
#method 1
new_x= 0
prediction = linear_regression(new_x)
print(f"Prediction for x={new_x}:{prediction}")
#method 2
new_x =pd.DataFrame({'YearsExperience':[0]})
print(model.predict(new_x))
```
